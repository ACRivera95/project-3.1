```python
def get_id_url(id):
    if id < 10 : 
        url_id = "00" + str(id) 
        return url_id
    elif id < 100:
        url_id = "0" + str(id) 
        return url_id
    else :
        url_id = str(id)
        return url_id
        
        
```

```python
def image_url(id):
    url = "https://assets.pokemon.com/assets/cms2/img/pokedex/full/{}.png".format(id)
    return url
```

```python
df = df.assign(Id= df["ID"].apply(get_id_url))
df.head()
```

```python
df['Id'].value_counts().idxmax()
```

```python
df[df['Id'].str.contains('479')]['Id']
```

```python
for index, row in df.iterrows():
    if index > 1:
        current_row = row
        previous_row_5th = df.iloc[index - 5]
        previous_row_4th = df.iloc[index - 4]
        previous_row_3th = df.iloc[index - 3]
        previous_row_2nd = df.iloc[index - 2]
        previous_row_1st = df.iloc[index - 1]
        
        if current_row["Id"] == previous_row_5th["Id"]:
            df.at[index, "Id"] = current_row["Id"] + "_f6"
        elif current_row["Id"] == previous_row_4th["Id"]:
            df.at[index, "Id"] = current_row["Id"] + "_f5"
        elif current_row["Id"] == previous_row_3th["Id"]:
            df.at[index, "Id"] = current_row["Id"] + "_f4"
        elif current_row["Id"] == previous_row_2nd["Id"]:
            df.at[index, "Id"] = current_row["Id"] + "_f3"
        elif current_row["Id"] == previous_row_1st["Id"]:
            df.at[index, "Id"] = current_row["Id"] + "_f2"
            

```

```python
df = df.assign(ImageUrl=df["Id"].apply(image_url))
df.head(10)
```

```python
def get_gen(id):
    if id < 152:
        generation = "1st"
        return generation
    elif id < 252:
        generation = "2nd"
        return generation
    elif id < 387:
        generation = "3rd"
        return generation
    elif id < 494:
        generation = "4th"
        return generation
    elif id < 650:
        generation = "5th"
        return generation
    elif id < 722:
        generation = "6th"
        return generation
    elif id < 810:
        generation = "7th"
        return generation
    elif id < 906:
        generation = "8th"
        return generation
    else:
        generation = "9th"
        return generation
```

```python
df = df.assign(Generation=df["ID"].apply(get_gen))
```

```python
df = df.assign(ID=df["ID"].apply(get_id_url))
df.head(10)
```
